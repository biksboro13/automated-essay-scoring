````python
# Load data

essay_data = pd.read_csv("domain123.csv")
#essay_data = essay_data.dropna(axis=1)
essay_data = essay_data.drop('rater1_domain1',1)
essay_data = essay_data.drop('rater2_domain1',1)
essay_data = essay_data.drop('rater3_domain1',1)

essay_data = essay_data.drop('rater1_trait1',1)
essay_data = essay_data.drop('rater1_trait2',1)
essay_data = essay_data.drop('rater1_trait3',1)
essay_data = essay_data.drop('rater1_trait4',1)
essay_data = essay_data.drop('rater1_trait5',1)
essay_data = essay_data.drop('rater1_trait6',1)

essay_data = essay_data.drop('rater2_trait1',1)
essay_data = essay_data.drop('rater2_trait2',1)
essay_data = essay_data.drop('rater2_trait3',1)
essay_data = essay_data.drop('rater2_trait4',1)
essay_data = essay_data.drop('rater2_trait5',1)
essay_data = essay_data.drop('rater2_trait6',1)

essay_data = essay_data.drop('rater3_trait1',1)
essay_data = essay_data.drop('rater3_trait2',1)
essay_data = essay_data.drop('rater3_trait3',1)
essay_data = essay_data.drop('rater3_trait4',1)
essay_data = essay_data.drop('rater3_trait5',1)
essay_data = essay_data.drop('rater3_trait6',1)

essay_data = essay_data.drop('rater1_domain2',1)
essay_data = essay_data.drop('rater2_domain2',1)
essay_data = essay_data.drop('domain2_score',1)

essay_data = essay_data.drop('essay_id',1)

print(essay_data.columns)
print("Length of the data=",len(essay_data))
````

**Data preprocessing**

````python
essay_data['essay'][11]
````


**As observed here, majority of the preprocessing has already been done in the data-set. The external referenecs are replaced with "@Text" label. These labels will not be considered for feature generation, ```helper.featureSet2()``` ensures to skip these labels.**

**Generating the feature columns**

````python
#Generating feature columns.
feature_data = helper.preprocess_data2feature(essay_data,'domain1_score')
feature_data.head()
````

**Split the data:**  
1. X_all : Contains the input essay and corresponding heuristic feature columns.  
2. y_all : The target value (essay grade) for the input essay.

````python
feature_cols = list(feature_data.columns[:-1])
X_all = feature_data[feature_cols]
y_all = feature_data['target'] #domain1_score is equal to sum of rater_domain1 and rater_domain2, so that columns are not needed
````

**Generating the features**  
The ```helper.GenerateFeatures()``` function will be used to generate the 14 heuristic features.  
List of feature columns: 

```feature_columns = ["essay","word_count","long_word_count","avg_word_length_per_essay",
"wrong_words","no_of_domain_words","word_to_sent_ratio","num_of_characters",
"sentence_count","noun_count","verb_count","comma_count","punctuation_count","adjective_count","adverb_count",
"quotation_mark_count"]```


````python
#Generate the feature set.
print len(X_all)
helper.GenerateFeatures(X_all)
print "Done"
````

**Saving the feature set**  
Clearly, genearting the feature set is a time consuming process, so the generated features will be saved so that the data visualization, model learning and testing can be carried out in a seperate notebook.  
1. Data visualization : Data_Exploration.ipynb
2. Model building and testing : Linear Regression model.ipynb

**NOTE: ** A NumPy array of feature vectors (consisting only of the features and not the essay) is also saved, these features may be used directly to combine with the Word Vectors generated by the word2vec model.

````python
# Save the features generated to a .csv file in case of failure or recovery purpose.
X_all_feature_vec = np.array(X_all.drop('essay',axis=1))
X_all.to_csv('model_and_visualization/features_set_rel3.csv')
y_all.to_csv('model_and_visualization/target_set_rel3.csv')

np.save('model_and_visualization/custom-feature-vectors.npy',X_all_feature_vec)
````
